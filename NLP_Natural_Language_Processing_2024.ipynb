**Tokenization**

```python
!pip install nltk
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
tok_q0=word_tokenize("what is the price of Bitcoin I am going to home")
tok_q0
```

```python
tok_q1=word_tokenize("Has the United States become the largest dictatorship in the world?")
tok_q1
```

**Stop Word Removal - Removing commonly occuring words**


```python
nltk.download('stopwords')
stop_words=nltk.corpus.stopwords.words('english')
",".join(stop_words)
```

```python
#Remove stop word
filtered_q0=[t for t in tok_q0 if t not in stop_words]
filtered_q0
```

**Stemming-go,going,gone->go; birds,bird->bird**

```python
nltk.download('wordnet')
ps=nltk.PorterStemmer()
stem_q0=[ps.stem(tok) for tok in filtered_q0]
stem_q0
```

**Stemming with second method**

```python
from nltk.stem.snowball import SnowballStemmer
stemmer=SnowballStemmer(language='english')
#stemmer.stem('going')
data=["i","am","goings","scholing"]
st=[stemmer.stem(tok) for tok in data]
st
```

**Lemmatization**
also corrects the data

```python
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
filt=[lemma.lemmatize(tok) for tok in data]
filt
```
