```python
#assign Male=1, Female=0
df['Sex'].unique()
```

```python
df=pd.get_dummies(df,columns=['Sex'],drop_first=True)
df.rename(columns={'Sex_Male':'Gender'},inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df['Gender']=df['Gender'].astype(int)
df.dtypes
```

```python
df.head()
```

```python
df.drop(['Male=1'],axis=1,inplace=True)
df.head()
```

```python
df['Cholesterol'].unique()
```

```python
df['Blood Pressure'].unique()
```

```python
def blood_pressure1(val):
  val=str(val)
  val=val.split('/')
  return int(val[0])

def blood_pressure2(val):
  val=str(val)
  val=val.split('/')
  return int(val[1])

df['Systolic_pressure']=df['Blood Pressure'].apply(blood_pressure1)
df['Diastolic_pressure']=df['Blood Pressure'].apply(blood_pressure2)
df.head()
```

```python
df.drop(['Blood Pressure'],axis=1,inplace=True)
df.head()
```

```python
df['Heart Rate'].isnull().sum()
```

```python
import warnings
warnings.filterwarnings('ignore')
df['Heart Rate'].isna().sum()
```

```python
df['Diet'].unique()
```

```python
def handle_Diet(value):
  value=str(value)
  if value=='Unhealthy':
    return 0
  elif value=='Average':
    return 1
  elif value=='Healthy':
    return 2
  else:
    return np.nan

df['Diet']=df['Diet'].apply(handle_Diet)
df.head()
```
