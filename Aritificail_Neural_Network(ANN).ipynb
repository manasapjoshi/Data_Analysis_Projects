**Artificial Neural Network **(ANN)

An artificial neural network (ANN) is a mathematical model that mimics the human brain's neurons and is used for solving problems. ANNs are made up of software modules called nodes, which are organized into layers and connected to each other.

```python
#get categorical variables
obj_col=df.select_dtypes(include='object')
print(obj_col.columns)
print(obj_col)
```

```python
#list of categorical
ct=(df.dtypes=='object')
cat_col=ct[ct].index
print(cat_col)
```

```python
#missing values in categorical data
for a in cat_col:
  print(a,df[a].isnull().sum())
```

```python
#fill the null values by mode into categorical data
for a in cat_col:
  df[a].fillna(df[a].mode()[0],inplace=True)
```

```python
#checking for null values again
for a in cat_col:
  print(a,df[a].isnull().sum())
```

```python
#get list of numeric columns
num_col=df.select_dtypes(include=['float64'])
print(num_col.columns)
```

```python
#show missing values
for a in num_col:
  print(a,df[a].isnull().sum())
```

```python
#fill missing values with median
for a in num_col:
  df[a].fillna(df[a].median(),inplace=True)
```

```python
#check for missing values again
for a in num_col:
  print(a,df[a].isnull().sum())
```

```python
#group by Rainfall by year wise
df['year']=pd.DatetimeIndex(df['Date']).year

#groupby rainfall according to year
gd=df.groupby('year')['Rainfall'].sum().reset_index()
gd
```

```python
#Visualization of year wise rain
plt.figure(figsize=(10,6))
sns.barplot(x='year',y='Rainfall',data=gd)
plt.show()
```

```python
#visualize with line plot using line seaborn
sns.lineplot(x='year',y='Rainfall',data=gd)
```

```python
#visualize with scatter plot
sns.scatterplot(x='year',y='Rainfall',data=gd)
```

```python
#apply label encoding for categorical columns
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for a in cat_col:
  df[a]=le.fit_transform(df[a])
  print(a,df[a].unique())
```

```python
#Drop Some Columns
features=df.drop(['RainTomorrow','Date','year'],axis=1)

#Create Target
target=df['RainTomorrow']

#Set the scaling features
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
features=sc.fit_transform(features)
features=pd.DataFrame(features)
features.head()
```

```python
#visualize using boxenplot chart the data
import seaborn as sns
sns.boxenplot(data=features)
plt.xticks(rotation=90)
plt.show()
```

```python
#Model Building
from sklearn.model_selection import train_test_split
x=df.drop(['RainTomorrow'],axis=1)
y=df['RainTomorrow']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#print train and test data
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
#
```

```python
#initializing the nm
from keras.models import Sequential
from keras.layers import Dense
model=Sequential() #initializing the neural network

#dense layers
model.add(Dense(32,activation='relu',input_dim=x_train.shape[1]))
model.add(Dense(64,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

#compiling the ANN
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

#train the ANN
model.fit(x_train,y_train,batch_size=32,epochs=50)
```

```python
#plotting training and validation loss over epochs
import matplotlib.pyplot as plt
plt.plot(model.history.history['loss'])
```

```python
#plotting training and validation accuracy over epochs
import matplotlib.pyplot as plt
plt.plot(model.history.history['accuracy'])
```

```python
#conclusion
#predicting the test set results and plotting
y_pred=model.predict(x_test)
y_pred=(y_pred>0.5)
print(y_pred)

```

```python
#plotting the prediction
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```
